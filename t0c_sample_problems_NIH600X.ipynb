## Example Graphs Using Pyplot


```python
t= np.arange(0., 2., 0.2) #evenly samples time on .2 (200ms) intervals
y=np.sin(2*np.pi*t) #y=sin (2*pi*t)
plt.plot(t,y) #plot (x,y)

plt.xlabel('time (s)')
plt.ylabel ('signal')
plt.title ("Wow, This Graph is Choppy")
```

To make the graph look nicer, let's try a higher sampling rate. 


```python
t=t= np.arange(0., 2., 0.02) #evenly samples time on .02 s intervals--> higher sampling rate
y=np.sin(2*np.pi*t) #y=sin (2*pi*t)
plt.plot(t,y) #plot (x,y)

plt.xlabel('time (s)')
plt.ylabel ('signal')
plt.title ("This Graph is Much Nicer")
```

## Making Plots Interactive with Bokeh

```python
t=t= np.arange(0., 2., 0.02) #evenly samples time on .02 s intervals--> higher sampling rate
y=np.sin(2*np.pi*t) #y=sin (2*pi*t)
plt.plot(t,y) #plot (x,y)

plt.xlabel('time (s)')
plt.ylabel ('signal')
plt.title ("An Interactive Graph")
bokeh.plotting.show(bokeh.mpl.to_bokeh())# make your plot interactive with bokeh
```
