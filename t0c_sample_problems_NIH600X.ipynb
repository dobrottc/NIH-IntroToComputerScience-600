## Example Graphs Using Pyplot


```python
t= np.arange(0., 2., 0.2) #evenly samples time on .2 (200ms) intervals
y=np.sin(2*np.pi*t) #y=sin (2*pi*t)
plt.plot(t,y, '.-', markersize=10) #plot (x,y)


plt.xlabel('time (s)')
plt.ylabel ('signal')
plt.title ("Wow, This Graph is Choppy")
```

To make the graph look nicer, let's try a higher sampling rate. 


```python
t=t= np.arange(0., 2., 0.02) #evenly samples time on .02 s intervals--> higher sampling rate
y=np.sin(2*np.pi*t) #y=sin (2*pi*t)
plt.plot(t,y) #plot (x,y)

plt.xlabel('time (s)')
plt.ylabel ('signal')
plt.title ("Smoother plot")
```

**Comments: 1. Plot ** *To plot, use a code cell.  We will not use Bokeh for this plot.*



```python
xs = np.linspace(0,2,100)
figH = plt.figure(figsize=(3,3*0.75))  # use a small plot as there is not much data
plt.plot(xs, 2*xs+0)
```

----

**Comments: 2. Equation** *To show the equation, use a Markdown cell. The equation for the line is 
$$y=m x + b$$
Where $m$ is the slope, and $b$ is the vertical offset.

-----

**Comments: 3: function** *Use a code cell to show your function.  You'll learn more about functions in a few weeks.*


```python
def area_under_line(slope=1):
    x = 2.0
    peak = slope * x
    area = peak * x / 2
    return (area)
```

**Comments: 4. Run the function** *Again, a code cell, which shows the function calls you wrote and its output.*

```python
print (area_under_line(0.5))
print (area_under_line(1.0))
print (area_under_line(2.0))
print (area_under_line(10.0))
```
